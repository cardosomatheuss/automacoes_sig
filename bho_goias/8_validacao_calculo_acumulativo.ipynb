```python
import numpy as np

for idx, row in bho_mp_teste.iterrows():
    tributarios = bho_mp_teste.loc[bho_mp_teste['nodestino'] == row['noorigem']]

    soma_jan = tributarios['new_q95_jan_ls'].sum()
    soma_fev = tributarios['new_q95_fev_ls'].sum()
    soma_mar = tributarios['new_q95_mar_ls'].sum()
    soma_abr = tributarios['new_q95_abr_ls'].sum()
    soma_mai = tributarios['new_q95_mai_ls'].sum()
    soma_jun = tributarios['new_q95_jun_ls'].sum()
    soma_jul = tributarios['new_q95_jul_ls'].sum()
    soma_ago = tributarios['new_q95_ago_ls'].sum()
    soma_set = tributarios['new_q95_set_ls'].sum()
    soma_out = tributarios['new_q95_out_ls'].sum()
    soma_nov = tributarios['new_q95_nov_ls'].sum()
    soma_dez = tributarios['new_q95_dez_ls'].sum()


    jan = soma_jan + row['vaz_increment_jan_ls']
    fev = soma_fev + row['vaz_increment_fev_ls']
    mar = soma_mar + row['vaz_increment_mar_ls']
    abr = soma_abr + row['vaz_increment_abr_ls']
    mai = soma_mai + row['vaz_increment_mai_ls']
    jun = soma_jun + row['vaz_increment_jun_ls']
    jul = soma_jul + row['vaz_increment_jul_ls']
    ago = soma_ago + row['vaz_increment_ago_ls']
    set = soma_set + row['vaz_increment_set_ls']
    out = soma_out + row['vaz_increment_out_ls']
    nov = soma_nov + row['vaz_increment_nov_ls']
    dez = soma_dez + row['vaz_increment_dez_ls']

    bho_mp_teste.at[idx,'verificacao_jan'] = np.isclose(jan,bho_mp_teste.at[idx,'new_q95_jan_ls'])
    bho_mp_teste.at[idx,'verificacao_fev'] = np.isclose(fev,bho_mp_teste.at[idx,'new_q95_fev_ls'])
    bho_mp_teste.at[idx,'verificacao_mar'] = np.isclose(mar,bho_mp_teste.at[idx,'new_q95_mar_ls'])
    bho_mp_teste.at[idx,'verificacao_abr'] = np.isclose(abr,bho_mp_teste.at[idx,'new_q95_abr_ls'])
    bho_mp_teste.at[idx,'verificacao_mai'] = np.isclose(mai,bho_mp_teste.at[idx,'new_q95_mai_ls'])
    bho_mp_teste.at[idx,'verificacao_jun'] = np.isclose(jun,bho_mp_teste.at[idx,'new_q95_jun_ls'])
    bho_mp_teste.at[idx,'verificacao_jul'] = np.isclose(jul,bho_mp_teste.at[idx,'new_q95_jul_ls'])
    bho_mp_teste.at[idx,'verificacao_ago'] = np.isclose(ago,bho_mp_teste.at[idx,'new_q95_ago_ls'])
    bho_mp_teste.at[idx,'verificacao_set'] = np.isclose(set,bho_mp_teste.at[idx,'new_q95_set_ls'])
    bho_mp_teste.at[idx,'verificacao_out'] = np.isclose(out,bho_mp_teste.at[idx,'new_q95_out_ls'])
    bho_mp_teste.at[idx,'verificacao_nov'] = np.isclose(nov,bho_mp_teste.at[idx,'new_q95_nov_ls'])
    bho_mp_teste.at[idx,'verificacao_dez'] = np.isclose(dez,bho_mp_teste.at[idx,'new_q95_dez_ls'])
        
```
