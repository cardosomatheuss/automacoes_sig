```python
import pandas as pd 
import geopandas as gpd 
```

```python
bho_saomarcos_ssd = gpd.read_file('D:\\matheus.cgomes\\Documentos\\GitHub\\gis_automations\\data\\bho_ana_sao_marcos_corrigido_22_05_2024.gpkg')
bho_nova_goias = gpd.read_file('D:\\matheus.cgomes\\Documentos\\GitHub\\gis_automations\\data\\bho_q95_qmed_correc_03-06-2024_final_v3_FINAL.gpkg')
```

```python
def correcao_saomarcos(bho_saomarcos_ssd,bho_goias_semad):
    bho_saomarcos_ssd['q95_jan_ls'] = bho_saomarcos_ssd['vaz_nu_qref1'] * 1000
    bho_saomarcos_ssd['q95_fev_ls'] = bho_saomarcos_ssd['vaz_nu_qref2'] * 1000
    bho_saomarcos_ssd['q95_mar_ls'] = bho_saomarcos_ssd['vaz_nu_qref3'] * 1000
    bho_saomarcos_ssd['q95_abr_ls'] = bho_saomarcos_ssd['vaz_nu_qref4'] * 1000
    bho_saomarcos_ssd['q95_mai_ls'] = bho_saomarcos_ssd['vaz_nu_qref5'] * 1000
    bho_saomarcos_ssd['q95_jun_ls'] = bho_saomarcos_ssd['vaz_nu_qref6'] * 1000
    bho_saomarcos_ssd['q95_jul_ls'] = bho_saomarcos_ssd['vaz_nu_qref7'] * 1000
    bho_saomarcos_ssd['q95_ago_ls'] = bho_saomarcos_ssd['vaz_nu_qref8'] * 1000
    bho_saomarcos_ssd['q95_set_ls'] = bho_saomarcos_ssd['vaz_nu_qref9'] * 1000
    bho_saomarcos_ssd['q95_out_ls'] = bho_saomarcos_ssd['vaz_nu_qref10'] * 1000
    bho_saomarcos_ssd['q95_nov_ls'] = bho_saomarcos_ssd['vaz_nu_qref11'] * 1000
    bho_saomarcos_ssd['q95_dez_ls'] = bho_saomarcos_ssd['vaz_nu_qref12'] * 1000

    # Cria um dicionário com os novos valores
    janeiro_valores = dict(zip(bho_saomarcos_ssd['tdr_cd_ottobacia'], bho_saomarcos_ssd['q95_jan_ls']))
    fevereiro_valores = dict(zip(bho_saomarcos_ssd['tdr_cd_ottobacia'], bho_saomarcos_ssd['q95_fev_ls']))
    marco_valores = dict(zip(bho_saomarcos_ssd['tdr_cd_ottobacia'], bho_saomarcos_ssd['q95_mar_ls']))
    abril_valores = dict(zip(bho_saomarcos_ssd['tdr_cd_ottobacia'], bho_saomarcos_ssd['q95_abr_ls']))
    maio_valores = dict(zip(bho_saomarcos_ssd['tdr_cd_ottobacia'], bho_saomarcos_ssd['q95_mai_ls']))
    junho_valores = dict(zip(bho_saomarcos_ssd['tdr_cd_ottobacia'], bho_saomarcos_ssd['q95_jun_ls']))
    julho_valores = dict(zip(bho_saomarcos_ssd['tdr_cd_ottobacia'], bho_saomarcos_ssd['q95_jul_ls']))
    agosto_valores = dict(zip(bho_saomarcos_ssd['tdr_cd_ottobacia'], bho_saomarcos_ssd['q95_ago_ls']))
    setembro_valores = dict(zip(bho_saomarcos_ssd['tdr_cd_ottobacia'], bho_saomarcos_ssd['q95_set_ls']))
    outubro_valores = dict(zip(bho_saomarcos_ssd['tdr_cd_ottobacia'], bho_saomarcos_ssd['q95_out_ls']))
    novembro_valores = dict(zip(bho_saomarcos_ssd['tdr_cd_ottobacia'], bho_saomarcos_ssd['q95_nov_ls']))
    dezembro_valores = dict(zip(bho_saomarcos_ssd['tdr_cd_ottobacia'], bho_saomarcos_ssd['q95_dez_ls']))

    # Atualiza os valores com base no dicionário
    bho_goias_semad['new_q95_jan_ls'] = bho_goias_semad['cobacia'].map(janeiro_valores).fillna(bho_goias_semad['new_q95_jan_ls'])
    bho_goias_semad['new_q95_fev_ls'] = bho_goias_semad['cobacia'].map(fevereiro_valores).fillna(bho_goias_semad['new_q95_fev_ls'])
    bho_goias_semad['new_q95_mar_ls'] = bho_goias_semad['cobacia'].map(marco_valores).fillna(bho_goias_semad['new_q95_mar_ls'])
    bho_goias_semad['new_q95_abr_ls'] = bho_goias_semad['cobacia'].map(abril_valores).fillna(bho_goias_semad['new_q95_abr_ls'])
    bho_goias_semad['new_q95_mai_ls'] = bho_goias_semad['cobacia'].map(maio_valores).fillna(bho_goias_semad['new_q95_mai_ls'])
    bho_goias_semad['new_q95_jun_ls'] = bho_goias_semad['cobacia'].map(junho_valores).fillna(bho_goias_semad['new_q95_jun_ls'])
    bho_goias_semad['new_q95_jul_ls'] = bho_goias_semad['cobacia'].map(julho_valores).fillna(bho_goias_semad['new_q95_jul_ls'])
    bho_goias_semad['new_q95_ago_ls'] = bho_goias_semad['cobacia'].map(agosto_valores).fillna(bho_goias_semad['new_q95_ago_ls'])
    bho_goias_semad['new_q95_set_ls'] = bho_goias_semad['cobacia'].map(setembro_valores).fillna(bho_goias_semad['new_q95_set_ls'])
    bho_goias_semad['new_q95_out_ls'] = bho_goias_semad['cobacia'].map(outubro_valores).fillna(bho_goias_semad['new_q95_out_ls'])
    bho_goias_semad['new_q95_nov_ls'] = bho_goias_semad['cobacia'].map(novembro_valores).fillna(bho_goias_semad['new_q95_nov_ls'])
    bho_goias_semad['new_q95_dez_ls'] = bho_goias_semad['cobacia'].map(dezembro_valores).fillna(bho_goias_semad['new_q95_dez_ls'])

    return bho_goias_semad
```

```python
bho_corrigida = correcao_saomarcos(bho_saomarcos_ssd,bho_nova_goias)
bho_corrigida
```

```python
bho_corrigida.to_file('bho_q95_qmed_03-06-2024_altosaomarcos_ok_FINAL.gpkg')
```
