```python
import pandas as pd
import geopandas as gpd
```

```python
estacoes = gpd.read_file('D:\\matheus.cgomes\\Documentos\\GitHub\\gis_automations\\data\\69_estacoes_vazoes.gpkg')
trechos = gpd.read_file('D:\\matheus.cgomes\\Documentos\\GitHub\\gis_automations\\data\\bho_q95_qmed_correc_29-05-2024_final_poligonos.gpkg')

```

```python
sjoin_estacoes = gpd.sjoin(estacoes,trechos,how='inner')
```

```python
columns_to_float = ['q95_jan_(m³/s)', 'q95_fev_(m³/s)', 'q95_mar_(m³/s)',
       'q95_abr_(m³/s)', 'q95_mai_(m³/s)', 'q95_jun_(m³/s)', 'q95_jul_(m³/s)',
       'q95_ago_(m³/s)', 'q95_set_(m³/s)', 'q95_out_(m³/s)', 'q95_nov_(m³/s)',
       'q95_dez_(m³/s)', 'qmed_jan_(m³/s)', 'qmed_fev_(m³/s)',
       'qmed_mar_(m³/s)', 'qmed_abr_(m³/s)', 'qmed_mai_(m³/s)',
       'qmed_jun_(m³/s)', 'qmed_jul_(m³/s)', 'qmed_ago_(m³/s)',
       'qmed_set_(m³/s)', 'qmed_out_(m³/s)', 'qmed_nov_(m³/s)',
       'qmed_dez_(m³/s)']
for column in columns_to_float:
    sjoin_estacoes[column] = sjoin_estacoes[column].str.replace(',','.').astype(float)

sjoin_estacoes['adkm2'] = sjoin_estacoes['adkm2'].astype(int)
```

```python
sjoin_estacoes['q95_jan_simulado'] = (sjoin_estacoes['new_q95_jan_ls']/sjoin_estacoes['nuareamont']) * sjoin_estacoes['adkm2']
sjoin_estacoes['q95_fev_simulado'] = (sjoin_estacoes['new_q95_fev_ls']/sjoin_estacoes['nuareamont']) * sjoin_estacoes['adkm2']
sjoin_estacoes['q95_mar_simulado'] = (sjoin_estacoes['new_q95_mar_ls']/sjoin_estacoes['nuareamont']) * sjoin_estacoes['adkm2']
sjoin_estacoes['q95_abr_simulado'] = (sjoin_estacoes['new_q95_abr_ls']/sjoin_estacoes['nuareamont']) * sjoin_estacoes['adkm2']
sjoin_estacoes['q95_mai_simulado'] = (sjoin_estacoes['new_q95_mai_ls']/sjoin_estacoes['nuareamont']) * sjoin_estacoes['adkm2']
sjoin_estacoes['q95_jun_simulado'] = (sjoin_estacoes['new_q95_jun_ls']/sjoin_estacoes['nuareamont']) * sjoin_estacoes['adkm2']
sjoin_estacoes['q95_jul_simulado'] = (sjoin_estacoes['new_q95_jul_ls']/sjoin_estacoes['nuareamont']) * sjoin_estacoes['adkm2']
sjoin_estacoes['q95_ago_simulado'] = (sjoin_estacoes['new_q95_ago_ls']/sjoin_estacoes['nuareamont']) * sjoin_estacoes['adkm2']
sjoin_estacoes['q95_set_simulado'] = (sjoin_estacoes['new_q95_set_ls']/sjoin_estacoes['nuareamont']) * sjoin_estacoes['adkm2']
sjoin_estacoes['q95_out_simulado'] = (sjoin_estacoes['new_q95_out_ls']/sjoin_estacoes['nuareamont']) * sjoin_estacoes['adkm2']
sjoin_estacoes['q95_nov_simulado'] = (sjoin_estacoes['new_q95_nov_ls']/sjoin_estacoes['nuareamont']) * sjoin_estacoes['adkm2']
sjoin_estacoes['q95_dez_simulado'] = (sjoin_estacoes['new_q95_dez_ls']/sjoin_estacoes['nuareamont']) * sjoin_estacoes['adkm2']


sjoin_estacoes['qmed_jan_simulado'] = (sjoin_estacoes['new_qmed_jan_ls']/sjoin_estacoes['nuareamont']) * sjoin_estacoes['adkm2']
sjoin_estacoes['qmed_fev_simulado'] = (sjoin_estacoes['new_qmed_fev_ls']/sjoin_estacoes['nuareamont']) * sjoin_estacoes['adkm2']
sjoin_estacoes['qmed_mar_simulado'] = (sjoin_estacoes['new_qmed_mar_ls']/sjoin_estacoes['nuareamont']) * sjoin_estacoes['adkm2']
sjoin_estacoes['qmed_abr_simulado'] = (sjoin_estacoes['new_qmed_abr_ls']/sjoin_estacoes['nuareamont']) * sjoin_estacoes['adkm2']
sjoin_estacoes['qmed_mai_simulado'] = (sjoin_estacoes['new_qmed_mai_ls']/sjoin_estacoes['nuareamont']) * sjoin_estacoes['adkm2']
sjoin_estacoes['qmed_jun_simulado'] = (sjoin_estacoes['new_qmed_jun_ls']/sjoin_estacoes['nuareamont']) * sjoin_estacoes['adkm2']
sjoin_estacoes['qmed_jul_simulado'] = (sjoin_estacoes['new_qmed_jul_ls']/sjoin_estacoes['nuareamont']) * sjoin_estacoes['adkm2']
sjoin_estacoes['qmed_ago_simulado'] = (sjoin_estacoes['new_qmed_ago_ls']/sjoin_estacoes['nuareamont']) * sjoin_estacoes['adkm2']
sjoin_estacoes['qmed_set_simulado'] = (sjoin_estacoes['new_qmed_set_ls']/sjoin_estacoes['nuareamont']) * sjoin_estacoes['adkm2']
sjoin_estacoes['qmed_out_simulado'] = (sjoin_estacoes['new_qmed_out_ls']/sjoin_estacoes['nuareamont']) * sjoin_estacoes['adkm2']
sjoin_estacoes['qmed_nov_simulado'] = (sjoin_estacoes['new_qmed_nov_ls']/sjoin_estacoes['nuareamont']) * sjoin_estacoes['adkm2']
sjoin_estacoes['qmed_dez_simulado'] = (sjoin_estacoes['new_qmed_dez_ls']/sjoin_estacoes['nuareamont']) * sjoin_estacoes['adkm2']
```

```python
sjoin_estacoes['percent_dif_q95_jan'] = (((sjoin_estacoes['q95_jan_simulado']-(sjoin_estacoes['q95_jan_(m³/s)']*1000))/(sjoin_estacoes['q95_jan_(m³/s)']*1000))*100).round(2)
sjoin_estacoes['percent_dif_q95_fev'] = (((sjoin_estacoes['q95_fev_simulado']-(sjoin_estacoes['q95_fev_(m³/s)']*1000))/(sjoin_estacoes['q95_fev_(m³/s)']*1000))*100).round(2)
sjoin_estacoes['percent_dif_q95_mar'] = (((sjoin_estacoes['q95_mar_simulado']-(sjoin_estacoes['q95_mar_(m³/s)']*1000))/(sjoin_estacoes['q95_mar_(m³/s)']*1000))*100).round(2)
sjoin_estacoes['percent_dif_q95_abr'] = (((sjoin_estacoes['q95_abr_simulado']-(sjoin_estacoes['q95_abr_(m³/s)']*1000))/(sjoin_estacoes['q95_abr_(m³/s)']*1000))*100).round(2)
sjoin_estacoes['percent_dif_q95_mai'] = (((sjoin_estacoes['q95_mai_simulado']-(sjoin_estacoes['q95_mai_(m³/s)']*1000))/(sjoin_estacoes['q95_mai_(m³/s)']*1000))*100).round(2)
sjoin_estacoes['percent_dif_q95_jun'] = (((sjoin_estacoes['q95_jun_simulado']-(sjoin_estacoes['q95_jun_(m³/s)']*1000))/(sjoin_estacoes['q95_jun_(m³/s)']*1000))*100).round(2)
sjoin_estacoes['percent_dif_q95_jul'] = (((sjoin_estacoes['q95_jul_simulado']-(sjoin_estacoes['q95_jul_(m³/s)']*1000))/(sjoin_estacoes['q95_jul_(m³/s)']*1000))*100).round(2)
sjoin_estacoes['percent_dif_q95_ago'] = (((sjoin_estacoes['q95_ago_simulado']-(sjoin_estacoes['q95_ago_(m³/s)']*1000))/(sjoin_estacoes['q95_ago_(m³/s)']*1000))*100).round(2)
sjoin_estacoes['percent_dif_q95_set'] = (((sjoin_estacoes['q95_set_simulado']-(sjoin_estacoes['q95_set_(m³/s)']*1000))/(sjoin_estacoes['q95_set_(m³/s)']*1000))*100).round(2)
sjoin_estacoes['percent_dif_q95_out'] = (((sjoin_estacoes['q95_out_simulado']-(sjoin_estacoes['q95_out_(m³/s)']*1000))/(sjoin_estacoes['q95_out_(m³/s)']*1000))*100).round(2)
sjoin_estacoes['percent_dif_q95_nov'] = (((sjoin_estacoes['q95_nov_simulado']-(sjoin_estacoes['q95_nov_(m³/s)']*1000))/(sjoin_estacoes['q95_nov_(m³/s)']*1000))*100).round(2)
sjoin_estacoes['percent_dif_q95_dez'] = (((sjoin_estacoes['q95_dez_simulado']-(sjoin_estacoes['q95_dez_(m³/s)']*1000))/(sjoin_estacoes['q95_dez_(m³/s)']*1000))*100).round(2)

sjoin_estacoes['percent_medio_dif_q95'] = ((sjoin_estacoes['percent_dif_q95_jan'].abs()+sjoin_estacoes['percent_dif_q95_fev'].abs()+sjoin_estacoes['percent_dif_q95_mar'].abs()+
                                           sjoin_estacoes['percent_dif_q95_abr'].abs()+sjoin_estacoes['percent_dif_q95_mai'].abs()+sjoin_estacoes['percent_dif_q95_jun'].abs()+
                                           sjoin_estacoes['percent_dif_q95_jul'].abs()+sjoin_estacoes['percent_dif_q95_ago'].abs()+sjoin_estacoes['percent_dif_q95_set'].abs()+
                                           sjoin_estacoes['percent_dif_q95_out'].abs()+sjoin_estacoes['percent_dif_q95_nov'].abs()+sjoin_estacoes['percent_dif_q95_dez'].abs())/12).round(2)




sjoin_estacoes['percent_dif_qmed_jan'] = (((sjoin_estacoes['qmed_jan_simulado']-(sjoin_estacoes['qmed_jan_(m³/s)']*1000))/(sjoin_estacoes['qmed_jan_(m³/s)']*1000))*100).round(2)
sjoin_estacoes['percent_dif_qmed_fev'] = (((sjoin_estacoes['qmed_fev_simulado']-(sjoin_estacoes['qmed_fev_(m³/s)']*1000))/(sjoin_estacoes['qmed_fev_(m³/s)']*1000))*100).round(2)
sjoin_estacoes['percent_dif_qmed_mar'] = (((sjoin_estacoes['qmed_mar_simulado']-(sjoin_estacoes['qmed_mar_(m³/s)']*1000))/(sjoin_estacoes['qmed_mar_(m³/s)']*1000))*100).round(2)
sjoin_estacoes['percent_dif_qmed_abr'] = (((sjoin_estacoes['qmed_abr_simulado']-(sjoin_estacoes['qmed_abr_(m³/s)']*1000))/(sjoin_estacoes['qmed_abr_(m³/s)']*1000))*100).round(2)
sjoin_estacoes['percent_dif_qmed_mai'] = (((sjoin_estacoes['qmed_mai_simulado']-(sjoin_estacoes['qmed_mai_(m³/s)']*1000))/(sjoin_estacoes['qmed_mai_(m³/s)']*1000))*100).round(2)
sjoin_estacoes['percent_dif_qmed_jun'] = (((sjoin_estacoes['qmed_jun_simulado']-(sjoin_estacoes['qmed_jun_(m³/s)']*1000))/(sjoin_estacoes['qmed_jun_(m³/s)']*1000))*100).round(2)
sjoin_estacoes['percent_dif_qmed_jul'] = (((sjoin_estacoes['qmed_jul_simulado']-(sjoin_estacoes['qmed_jul_(m³/s)']*1000))/(sjoin_estacoes['qmed_jul_(m³/s)']*1000))*100).round(2)
sjoin_estacoes['percent_dif_qmed_ago'] = (((sjoin_estacoes['qmed_ago_simulado']-(sjoin_estacoes['qmed_ago_(m³/s)']*1000))/(sjoin_estacoes['qmed_ago_(m³/s)']*1000))*100).round(2)
sjoin_estacoes['percent_dif_qmed_set'] = (((sjoin_estacoes['qmed_set_simulado']-(sjoin_estacoes['qmed_set_(m³/s)']*1000))/(sjoin_estacoes['qmed_set_(m³/s)']*1000))*100).round(2)
sjoin_estacoes['percent_dif_qmed_out'] = (((sjoin_estacoes['qmed_out_simulado']-(sjoin_estacoes['qmed_out_(m³/s)']*1000))/(sjoin_estacoes['qmed_out_(m³/s)']*1000))*100).round(2)
sjoin_estacoes['percent_dif_qmed_nov'] = (((sjoin_estacoes['qmed_nov_simulado']-(sjoin_estacoes['qmed_nov_(m³/s)']*1000))/(sjoin_estacoes['qmed_nov_(m³/s)']*1000))*100).round(2)
sjoin_estacoes['percent_dif_qmed_dez'] = (((sjoin_estacoes['qmed_dez_simulado']-(sjoin_estacoes['qmed_dez_(m³/s)']*1000))/(sjoin_estacoes['qmed_dez_(m³/s)']*1000))*100).round(2)


sjoin_estacoes['percent_medio_dif_qmed'] = ((sjoin_estacoes['percent_dif_qmed_jan'].abs()+sjoin_estacoes['percent_dif_qmed_fev'].abs()+sjoin_estacoes['percent_dif_qmed_mar'].abs()+
                                           sjoin_estacoes['percent_dif_qmed_abr'].abs()+sjoin_estacoes['percent_dif_qmed_mai'].abs()+sjoin_estacoes['percent_dif_qmed_jun'].abs()+
                                           sjoin_estacoes['percent_dif_qmed_jul'].abs()+sjoin_estacoes['percent_dif_qmed_ago'].abs()+sjoin_estacoes['percent_dif_qmed_set'].abs()+
                                           sjoin_estacoes['percent_dif_qmed_out'].abs()+sjoin_estacoes['percent_dif_qmed_nov'].abs()+sjoin_estacoes['percent_dif_qmed_dez'].abs())/12).round(2)




```
