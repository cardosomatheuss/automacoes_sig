```python
import pandas as pd
import geopandas as gpd
```

```python
estacoes_points = gpd.read_file('D:\\matheus.cgomes\\Documentos\\GitHub\\gis_automations\\data\\69_estacoes_bho_goias.gpkg')
estacoes_vazoes = pd.read_csv('D:\\matheus.cgomes\\Documentos\\GitHub\\gis_automations\\data\\Estacoes_FLU_BHO_Goias_V2.xlsx - Cópia de Estações versão final.csv',sep=',')
```

```python
def transpose_dataset(df,first_column = str):
    df_transposed = df.set_index(first_column).unstack().unstack().reset_index(names=first_column).rename_axis(columns=None)
    return df_transposed

estacoes_vazoes_ok = transpose_dataset(estacoes_vazoes,'codigo')
estacoes_vazoes_ok
```

```python
estacoes_points_filtered = estacoes_points[['id', 'codigo', 'codalt', 'nome', 'rio','geometry']]
estacoes_vazoes_ok_filtered = estacoes_vazoes_ok[['codigo', 'adkm2', 'anoinicio', 'anofim', 'anototal','q95_jan_(m³/s)',
       'q95_fev_(m³/s)', 'q95_mar_(m³/s)', 'q95_abr_(m³/s)', 'q95_mai_(m³/s)',
       'q95_jun_(m³/s)', 'q95_jul_(m³/s)', 'q95_ago_(m³/s)', 'q95_set_(m³/s)',
       'q95_out_(m³/s)', 'q95_nov_(m³/s)', 'q95_dez_(m³/s)','qmed_jan_(m³/s)',
       'qmed_fev_(m³/s)', 'qmed_mar_(m³/s)', 'qmed_abr_(m³/s)',
       'qmed_mai_(m³/s)', 'qmed_jun_(m³/s)', 'qmed_jul_(m³/s)',
       'qmed_ago_(m³/s)', 'qmed_set_(m³/s)', 'qmed_out_(m³/s)',
       'qmed_nov_(m³/s)', 'qmed_dez_(m³/s)']]

merge = pd.merge(estacoes_points_filtered,estacoes_vazoes_ok_filtered,on='codigo')
```

```python
merge_gdf = gpd.GeoDataFrame(merge,geometry='geometry')

merge_gdf.to_file('D:\\matheus.cgomes\\Documentos\\GitHub\\gis_automations\\data\\69_estacoes_vazoes.gpkg')
```
