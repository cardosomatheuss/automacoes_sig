```python
def union_areas_incrementais_centros_superf(centros_superficie,areas_incrementais):
    union2_qmed_points = pd.merge(centros_superficie,areas_incrementais[['codigo','qmed_esp_jan_(l/s/km²)',
       'qmed_esp_fev_(l/s/km²)', 'qmed_esp_mar_(l/s/km²)',
       'qmed_esp_abr_(l/s/km²)', 'qmed_esp_mai_(l/s/km²)',
       'qmed_esp_jun_(l/s/km²)', 'qmed_esp_jul_(l/s/km²)',
       'qmed_esp_ago_(l/s/km²)', 'qmed_esp_set_(l/s/km²)',
       'qmed_esp_out_(l/s/km²)', 'qmed_esp_nov_(l/s/km²)',
       'qmed_esp_dez_(l/s/km²)','q95esp_jan', 'q95esp_fev', 'q95esp_mar', 'q95esp_abr', 'q95esp_mai',
       'q95esp_jun', 'q95esp_jul', 'q95esp_ago', 'q95esp_set', 'q95esp_out',
       'q95esp_nov', 'q95esp_dez']],on='codigo',how='inner')
    
    columns_to_convert = ['q95esp_jan', 'q95esp_fev', 'q95esp_mar', 'q95esp_abr', 'q95esp_mai',
       'q95esp_jun', 'q95esp_jul', 'q95esp_ago', 'q95esp_set', 'q95esp_out',
       'q95esp_nov', 'q95esp_dez', 'qmed_esp_jan_(l/s/km²)',
       'qmed_esp_fev_(l/s/km²)', 'qmed_esp_mar_(l/s/km²)',
       'qmed_esp_abr_(l/s/km²)', 'qmed_esp_mai_(l/s/km²)',
       'qmed_esp_jun_(l/s/km²)', 'qmed_esp_jul_(l/s/km²)',
       'qmed_esp_ago_(l/s/km²)', 'qmed_esp_set_(l/s/km²)',
       'qmed_esp_out_(l/s/km²)', 'qmed_esp_nov_(l/s/km²)',
       'qmed_esp_dez_(l/s/km²)']
    
    for column in columns_to_convert:
         union2_qmed_points[column] = union2_qmed_points[column].str.replace(',','.').astype(float)

    return union2_qmed_points
```
